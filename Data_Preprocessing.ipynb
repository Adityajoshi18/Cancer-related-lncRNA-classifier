```python
import pandas as pd
```

```python
data=pd.read_csv("mid-sem/Data.csv")
```

```python
data.head()
```

```python
PositiveData=pd.read_csv("mid-sem/Positive.csv")
```

```python
NegativeData=pd.read_csv("mid-sem/Negative.csv")
```

```python
NegativeData.head()
```

```python
len(PositiveData)
```

```python
len(NegativeData)
```

```python
PositiveData = PositiveData.iloc[:,0].tolist()
```

```python
NegativeData = NegativeData.iloc[:,0].tolist()
```

```python
PositiveData
```

```python
NegativeData
```

```python
PositiveIndex=[]
for i in range(len(PositiveData)):
    index=data[data['Gene_ID']==PositiveData[i]].index.tolist()
    PositiveIndex.extend(index)
```

```python
PositiveIndex
```

```python
NegativeIndex=[]
for i in range(len(NegativeData)):
    index=data[data['Gene_ID']==NegativeData[i]].index.tolist()
    NegativeIndex.extend(index)
```

```python
UsefulIndex=PositiveIndex+NegativeIndex
TrainingData=data.iloc[UsefulIndex, :]
```

```python
TrainingData.head()
```

```python
len(TrainingData)
```

```python
ValidData=data.drop(UsefulIndex)
```

```python
ValidData.head()
```

```python
len(ValidData)
```

```python
ValidName=ValidData['Gene_ID']
```

```python
ValidData=ValidData.drop(['Gene_ID'], axis=1)
```

```python
ValidData.head()
```

```python
P_label=[1 for i in range(len(PositiveIndex))]
```

```python
N_label=[0 for i in range(len(NegativeIndex))]
```

```python
TrainingLabel=P_label+N_label
```

```python
TrainingData=TrainingData.assign(Label=TrainingLabel)
```

```python
TrainingData['Label']
```

```python
ShuffleSeed = 442
```

```python
TrainingData=TrainingData.sample(frac=1, random_state=ShuffleSeed)
```

```python
TrainingLabel=TrainingData['Label'].tolist()
```

```python
TrainingName=TrainingData['Gene_ID']
```

```python
TrainingData=TrainingData.drop(['Label', 'Gene_ID'], axis=1)
```

```python
TrainingData.head()
```

```python
#TrainingData, TrainingLabel, ValidData
```

```python
TrainingData=TrainingData.fillna(0)
```

```python
ValidData=ValidData.fillna(0)
```

```python
FeatureName=TrainingData.columns.values
```

```python
FeatureName
```

```python
from sklearn.feature_selection import VarianceThreshold
```

```python
selector=VarianceThreshold(threshold=1)
selector.fit(TrainingData)
RemainedFeatureLoc=selector.get_support(indices=True)
TrainingData=selector.transform(TrainingData)
ValidData=selector.transform(ValidData)
```

```python
RemainedFeatureLoc
```

```python
RemainedFeatureName=FeatureName[RemainedFeatureLoc]
```

```python
RemainedFeatureName
```

```python
TrainingSet1=pd.DataFrame(TrainingData, index=None, columns=RemainedFeatureName)
```

```python
TrainingSet1.to_csv("mid-sem/TrainingSet2.csv", index=None)
#TrainingSet1.to_csv("TrainingSet.csv", index=None)
```

```python
ValidSet1 = pd.DataFrame(ValidData, index=None, columns=RemainedFeatureName)
```

```python
ValidSet1.to_csv("mid-sem/ValidSet2.csv", index=None)
#ValidSet1.to_csv("ValidSet.csv", index=None)
```
